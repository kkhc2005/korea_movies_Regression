```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

movies_train_data = pd.read_csv('../korea_movies_box_off_num/movies_train.csv')
movies_test_data = pd.read_csv('../korea_movies_box_off_num/movies_test.csv')
submission_data = pd.read_csv('../korea_movies_box_off_num/submission.csv')
movies_train_data.head(10)
```

```python
movies_train_data['genre']=movies_train_data['genre'].map({'드라마':1,'멜로/로맨스':2,'다큐멘터리':3,'코미디':4})
movies_test_data['genre']=movies_test_data['genre'].map({'드라마':1,'멜로/로맨스':2,'다큐멘터리':3,'코미디':4})
```

```python
movies_train_data['screening_rat']=movies_train_data['screening_rat'].map({'청소년 관람불가':0,'12세 관람가':1,'15세 관람가':2,'전체 관람가':3})
movies_test_data['screening_rat']=movies_test_data['screening_rat'].map({'청소년 관람불가':0,'12세 관람가':1,'15세 관람가':2,'전체 관람가':3})
```

```python
X=movies_train_data[['genre', 'time', 'screening_rat', 'dir_prev_bfnum', 'dir_prev_num', 'num_staff', 'num_actor']]
y=movies_train_data['box_off_num']

X_test=movies_test_data[['genre', 'time', 'screening_rat', 'dir_prev_bfnum', 'dir_prev_num', 'num_staff', 'num_actor']]

X = X.fillna(-1)
X_test = X_test.fillna(-1)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(random_state=0)
```

```python
clf=clf.fit(X,y)
```

```python
movies_test_data['box_off_num']=clf.predict(X_test)
```

```python
movies_test_data[['title','box_off_num']].to_csv("submission_baseline.csv",index=False)
```
